```python
delhi_rest_data = pd.DataFrame()
delhi_rest_data_arr = []
for i in range(1,2560):
    webd.get("https://www.zomato.com/ncr/restaurants?page="+str(i))
    if webd.find_elements_by_css_selector('#orig-search-list .search-card'):
        search_cards = webd.find_elements_by_css_selector('#orig-search-list .search-card')
        for search_card in search_cards:
            delhi_rest_data_dict = {}
            try:
                if(search_card.find_element_by_css_selector('.result-title')):
                    delhi_rest_data_dict['Restaurent_name'] = search_card.find_element_by_css_selector('.result-title').text
                    try:
                        delhi_rest_data_dict['Restaurent_type'] = search_card.find_element_by_css_selector('.res-snippet-small-establishment').text
                    except:
                        print('no restaurant type')
                    try:
                        delhi_rest_data_dict['Restaurent_location'] = search_card.find_element_by_css_selector('.search_result_subzone').text
                    except:
                        print('no restaurant location')
                    try:
                        delhi_rest_data_dict['Restaurent_address'] = search_card.find_element_by_css_selector('.search-result-address').text
                    except:
                        print('no restaurant address')
                    try:
                        delhi_rest_data_dict['Restaurent_cost'] = search_card.find_element_by_css_selector('.res-cost span:nth-child(2)').text
                    except:
                        print('no restaurant cost')
                    try:
                        delhi_rest_data_dict['Restaurent_time'] = search_card.find_element_by_css_selector('.res-timing').text
                    except:
                        print('no restaurant time')
                    try:
                        delhi_rest_data_dict['Restaurent_cuisine'] = search_card.find_element_by_css_selector('.search-page-text div:nth-child(1) span:nth-child(2)').text
                    except:
                        print('no restaurant cuisine')
                    try:
                        #if search_card.find_element_by_css_selector('#orig-search-list .search-card .both-rating'):
                            #rest_dining_rev = search_card.find_element_by_css_selector('#orig-search-list  .search-card .both-rating')
                        rating_divs = search_card.find_elements_by_css_selector('#orig-search-list  .search-card .both-rating .rating-div')
                        for x in rating_divs:
                            starColor = x.find_element_by_css_selector('.star').value_of_css_property("color")
                            if(starColor == 'rgba(28, 28, 28, 1)'):
                                rating_value = x.find_element_by_css_selector('.rating-value').text.replace("\n","")
                                delhi_rest_data_dict['diner_rating'] = rating_value
                            else:
                                rating_value = x.find_element_by_css_selector('.rating-value').text.replace("\n","")
                                delhi_rest_data_dict['delivery_rating'] = rating_value
                    except Exception as e:
                        print(e)
                        print('no double rating')
                    try:
                        #elif webd.find_elements_by_css_selector('#orig-search-list  .search-card .single-rating'):
                        rest_dining_rev = search_card.find_element_by_css_selector('#orig-search-list   .search-card .single-rating')
                        starColor = rest_dining_rev.find_element_by_css_selector('.star').value_of_css_property("color")
                        if(starColor == 'rgba(28, 28, 28, 1)'):
                            rating_value = rest_dining_rev.find_element_by_css_selector('.rating-value').text.replace("\n","")
                            delhi_rest_data_dict['diner_rating'] = rating_value
                        else:
                                rating_value = rest_dining_rev.find_element_by_css_selector('.rating-value').text.replace("\n","")
                                delhi_rest_data_dict['delivery_rating'] = rating_value
                        #delhi_rest_data_dict['Restaurent_type'] = search_card.find_element_by_css_selector('.res-snippet-small-establishment').text
                    except Exception as e:
                        print(e)
                        print('no single rating')
                    delhi_rest_data_arr.append(delhi_rest_data_dict)
            except Exception as e:
                print(e)
                print('error in getting elements')
delhi_rest_data_arr
```

```python
delhi_rest_data = pd.DataFrame(delhi_rest_data_arr)
```
